```python
# in here agents represants the customers
from mesa import Agent, Model  # model represants the cinema hall's simulation environment
from mesa.time import RandomActivation #triggers when random customers attempt to book seats
from mesa.space import MultiGrid # seats with rows and columns in the cinema hall environment
from mesa.datacollection import DataCollector # collecting the data during the simmulation process
```

```python
class CinemaModel(Model):#inherits from the model class
    def __init__(self,N,width,height):#constructor of the class(N= Number of agents/customers, Width,height:- dimmensions/capacity of the cinmea hall)
        
        self.num_agents=N
        self.grid=MultiGrid(width,height,True)# arrange the cinmea hall space into rows /coumns with seating arrangements
        self.schedule=RandomActivation(self)#Randomly select agents
        for i in range(self.num_agents):#create agents
            a=CustomerAgent(i,self)# create unique customerAgent with id
            x=self.random.randrange(self.grid.width)# randomly selects a column index
            y=self.random.randrange(self.grid.height)# randomly selects a row index
            self.grid.place_agent(a,(x,y))# place the agent at the corresponding grid location
            self.schedule.add(a)# add the agent to the sheduler
            
        self.datacollector=DataCollector(
        agent_reporters={"Booking":"booking"}
        
        )# track the booking attribute of each agent
        
    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
class CustomerAgent(Agent):
    def __init__(self,unique_id,model):
        super().__init__(unique_id,model)
        self.name=f"Customer_{unique_id}"
        self.booking=0
    def step(self):
        x=self.random.randrange(self.model.grid.width)# randomly selects a column index
        y=self.random.randrange(self.model.grid.height)# randomly selects a row index
        
        cell_content=self.model.grid.get_cell_list_contents([(x,y)])#try to book a seat
        if len(cell_content)==0:# check whether the cell is occupied
            self.booking=1 # marks the seat as booked
            print(f"{self.name}successfully booked seat({x},{y})")
        else:
            self.booking=0
            print(f"{self.name}failed to booked seat({x},{y})")
```

```python
width=10
height=10
num_agents=15
model=CinemaModel(num_agents,width,height)

for i in range(20):
    model.step()
    
for agent in model.schedule.agents:
    print(f"Customer {agent.unique_id}booking status: {agent.booking}")
```
